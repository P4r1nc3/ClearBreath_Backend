```python
import os
import requests
import time
import matplotlib.pyplot as plt
```

```python
# Get token from environment variable
token = "eyJhbGciOiJIUzI1NiJ9.eyJzdWIiOiJrb25yYWQudHVwdGFAZ21haWwuY29tIiwiaWF0IjoxNzI5NDExNTMzLCJleHAiOjE3Mjk0MTI5NzN9.-AW_y1bZO4Bv5RcKx3X6dKr0KPsZTfyb6bR1so0oAgQ"

# Define the URL and headers with the token
url = "http://localhost:8080/pollution/lat/10/lng/20"
headers = {
    "Authorization": f"Bearer {token}"
}
```

```python
# Store response times
response_times = []

# Measure the total time taken for 1000 requests
total_start_time = time.time()

# Execute 1000 requests and measure response times
for _ in range(1000):
    start_time = time.time()
    response = requests.get(url, headers=headers)
    end_time = time.time()
    
    # Calculate response time in milliseconds
    response_time = (end_time - start_time) * 1000
    response_times.append(response_time)

# Measure total execution time
total_end_time = time.time()
total_execution_time = total_end_time - total_start_time  # in seconds
```

```python
# Plot the response times
plt.figure(figsize=(10, 6))
plt.plot(response_times, marker='o', linestyle='-', color='b')
plt.xlabel('Numer wysłanego zapytania')
plt.ylabel('Czas odpowiedzi (ms)')
plt.grid(True)
plt.show()

# Print the total execution time
print(f"Całkowity czas wykonania 1000 requestów: {total_execution_time:.2f} sekund")
```
